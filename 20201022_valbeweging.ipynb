## Theoretische benadering:

```
#  valbeweging: 
#  snelheid van een voorwerp dat losgelaten wordt op hoogte h wanneer het de grond bereikt. 

h = 10   # begin hoogte in meter
g = 9.81 # valversnelling (in m/s^2)

v_th = np.sqrt(2*g*h)

print('Theoretisch predictie:  v = ', v_th ,'m/s')
```

## Bepalen van de snelheid dmv simulatie
## Valbeweging - geen rekening houdend met de weerstand

```
#  SIMULATIE

y = h     #  begin hoogte
y_list = [h]

v = 0     #  begin snelheid 
dt = 0.001 

while y > 0: 

  # update de verticale snelheid
  v = v + g*dt

  # bereken de nieuwe y-waarde
  y = y - v*dt

  y_list.append(y)

#  Print de laatste waarde voor de snelheid: 

print(' ')
print('Resultaat voor de simulatie:  v = ', v, 'm/s')
```

```
fig, ax = plt.subplots(figsize=(10,8))
x = np.arange(0,len(y_list)*dt, dt)

ax.grid(True)
ax.plot(x, y_list , 'r-')
ax.set_title("Hoogte van het object - rekening houdend met de weerstand")
ax.set_xlabel("t")
ax.set_ylabel('h')
```

## Valbeweging rekening houdend met de luchtweerstand

```
#  SIMULATIE

#  De luchtweerstand kracht (en dus de versnelling) die een lichaam voelt bij een snelheid v 
#  is evenredig met die snelheid:  a_R = versnelling (eigenlijk: vertraging) door weerstand 
#  is daarom:   a_R = r*v  waarbij r de frictie coefficient is voor het lichaam 
#  (die ondermeer afhangt van de vorm)

y = h
y_list = [h]
v = 0   #  begin snelheid
r = 0.5    #  weerstands coefficient  
dt = 0.001

while y > 0: 
  # update van de versnelling (wijzigt nu omdat we rekening houden met de weerstand)
  a =  g - r*v
  # update de verticale snelheid
  v = v + a*dt

  # bereken de nieuwe y-waarde
  y = y - v*dt
  y_list.append(y)

#  Print de laatste waarde voor de snelheid: 

print(' ')
print('Resultaat voor de simulatie:  v = ', v, 'm/s')
print('Resultaat voor de simulatie:  v = ', a, 'm/s^2')
```

```
fig, ax = plt.subplots(figsize=(10,8))
x = np.arange(0,len(y_list)*dt, dt)

ax.grid(True)
ax.plot(x, y_list , 'r-')
ax.set_title("Hoogte van het object - rekening houdend met de weerstand")
ax.set_xlabel("t")
ax.set_ylabel('h')
```

- Er werken 2 krachten op de kogel: 
  - $\vec{F} = - k \vec{v}$ de wrijvingskracht en
  - $m \vec{g}$ de zwaartekracht


- De totale kracht die op de kogel inwerkt is $\vec{T} = \vec{F} + m\vec{g} $
- Delen we bovenstaande vergelijking door de massa $m$, dan krijgen we:
$$
\frac{\vec{T}}{m} = \frac{\vec{F}}{m} + \frac{m\vec{g}}{m}
$$
- Dan krijgen we:
$$
\vec{a} = -\rho \vec{v} + \vec{g}
$$
- We ontboncen de versnellingsvector nu in een $x$-componenten en een $y$-component:
$$
a_x = g_x -\rho v_x  = -\rho v_x 
$$
$$
a_y = g_y -\rho v_y = - g -\rho v_y 
$$
